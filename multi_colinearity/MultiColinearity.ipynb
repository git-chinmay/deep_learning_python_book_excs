```python
import pandas as pd
data = pd.read_csv('salary_data.csv')
data
```

```python
data.head()
```

```python
#!pip install statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(x):
    vif = pd.DataFrame()
    vif['varibales'] = x.columns
    vif['vif_score'] = [variance_inflation_factor(x, i)for i in range(x.shape[1])]
    return vif
```

```python
data_vif = calc_vif(data)
data_vif
```

```python
# Dropping the Years Service column and recalculating
data_vif = calc_vif(data.drop('Years Service', axis=1))
data_vif
```

```python
# Combining both the columns
new_data = data.copy()
new_data['Age at joining'] = new_data.apply(lambda x: x['Age'] - x['Years Service'], axis=1)
new_data.head()
```

```python
calc_vif(new_data.drop(['Age', 'Years Service'], axis=1))
```
